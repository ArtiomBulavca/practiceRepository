```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

lab3_data = sns.load_dataset('titanic')

lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'male'), 'who'] = 'boy'
lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'female'), 'who'] = 'girl'

lab3_data.loc[lab3_data['survived'] == 0, 'survived_str'] = 'no'
lab3_data.loc[lab3_data['survived'] == 1, 'survived_str'] = 'yes'

lab3_data = lab3_data[lab3_data['class'] != 'Second']

sns.set_style("white")
sns.set_context("paper")

ax = sns.boxplot(data=lab3_data, x='class', y='age', order=['Third', 'First'], color='white', fliersize=0)

ax.set(xlabel='Boarding Class', ylabel='Age')

plt.title('Age Distribution by Boarding Class')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

lab3_data = sns.load_dataset('titanic')

lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'male'), 'who'] = 'boy'
lab3_data.loc[(lab3_data['who'] == 'child') & (lab3_data['sex'] == 'female'), 'who'] = 'girl'

lab3_data.loc[lab3_data['survived'] == 0, 'survived_str'] = 'no'
lab3_data.loc[lab3_data['survived'] == 1, 'survived_str'] = 'yes'

sns.set_style("whitegrid")
sns.set_context("paper")

color_palette = {'man': 'blue', 'woman': 'tab:orange', 'boy': 'cornflowerblue', 'girl': 'sandybrown'}

plt.figure(figsize=(10, 6))
ax = sns.stripplot(data=lab3_data, x='survived_str', y='age', hue='who', dodge=True, palette=color_palette)

ax.set(xlabel='survived_str', ylabel='age')

plt.title('Age Distribution by Survival Status and Category')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

```python
fig, axes2 = plt.subplots(figsize=(10, 6))
sns.stripplot(data=lab3_data, x='survived_str', y='age', hue='who', jitter=0.3, alpha=0.7, dodge=True, ax=axes2)
sns.boxplot(data=lab3_data, x='survived_str', y='age', hue='who', ax=axes2, dodge=True, boxprops=dict(facecolor='white', alpha=1.0))  # Set boxprops to make boxplot fully visible and white

plt.xlabel('survived')
handles, labels = axes2.get_legend_handles_labels()
plt.legend(handles[4:8], labels[4:8], title='Who')

axes2.set_ylim(0, 80)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=lab3_data, x='survived_str', y='age', hue='who', palette='Set2')

plt.xlabel('Survived')
plt.ylabel('Age')

sns.despine()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')

sns.set_style("whitegrid")
sns.set_context("paper")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic_data, x='age', y='fare', hue='class', palette='muted', s=100)

plt.xlabel('Age')
plt.ylabel('Fare')

sns.despine()
plt.title('Age vs. Fare Colored by Passenger Class')

plt.show()
```

```python
lab3_data_subset=lab3_data[['class','who','fare']]
lab3_data_subset=lab3_data_subset.dropna(how='any')
lab3_data_subsetAverage=lab3_data_subset.groupby(['class','who']).mean()
lab3_data_subsetAverage=lab3_data_subsetAverage.reset_index()
lab3_data_subsetAverage.head()
```
